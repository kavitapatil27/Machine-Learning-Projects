```python
cols=df.iloc[:, [1, 2, 3, 4]].values
```

```python
cols
```

```python
#Finding the optimum number of clusters for k-means classification
from sklearn.cluster import KMeans
cluster_arr = []

for i in range(1, 10):
    means = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(cols)
    cluster_arr.append(kmeans.inertia_)
    
#Plotting the results onto a line graph
plt.plot(range(1, 10), cluster_arr)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('cluster sum of squares') 
plt.show()
```

From the above graph we realise that the optimum number of clusters is where the 'elbow' occurs. The above method is known as the 'elbow method'. If we observe the graph, the elbow occurs at 3.

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(cols)
```

```python
y_kmeans
```

```python
cols[y_kmeans == 0, 0]
```

```python
#Visualization of clusters
plt.figure(figsize=(10,8))
plt.scatter(cols[y_kmeans == 0, 0], cols[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(cols[y_kmeans == 1, 0], cols[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(cols[y_kmeans == 2, 0], cols[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Iris-virginica')

#Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'yellow', label = 'Centroids')

plt.legend()
```

The above graph shows that k=3 is the optimum value for clusters. Also, all the data points in the clusters are close to the centroid which makes k as the optimum value.
